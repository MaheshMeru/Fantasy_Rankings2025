```python
import requests
import json
from datetime import datetime


import pandas as pd


team_dict = {'Mahender': [{'Player': 'Ruturaj Gaikwad', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 17.5}, {'Player': 'Ravichandran Ashwin', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 5.25}, {'Player': 'Mukesh Choudhary', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 0.1}, {'Player': 'MS Dhoni', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 1.25}, {'Player': 'Anshul Kamboj', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 0.75}, {'Player': 'Tilak Varma', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 17.5}, {'Player': 'Mohit Sharma', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 0.75}, {'Player': 'Josh Inglis', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 3.5}, {'Player': 'Shashank Singh', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 8.25}, {'Player': 'Marcus Stoinis', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 9.0}, {'Player': 'Shahrukh Khan', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Shimron Hetmyer', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 2.0}, {'Player': 'Travis Head', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 20.0}, {'Player': 'Aniket Verma', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 1.4}, {'Player': 'Josh Hazlewood', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 2.25}, {'Player': 'Rishabh Pant', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 17.0}], 'Sai Kiran': [{'Player': 'Noor Ahmad', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 2.75}, {'Player': 'Devon Conway', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 9.5}, {'Player': 'Shreyas Gopal', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 0.45}, {'Player': 'Vijay Shankar', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Rahul Tripathi', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 5.25}, {'Player': 'Trent Boult', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 14.5}, {'Player': 'Deepak Chahar', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 4.25}, {'Player': 'Mitchell Santner', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 3.25}, {'Player': 'Faf du Plessis', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 5.5}, {'Player': 'Kuldeep Yadav', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 6.25}, {'Player': 'Yashasvi Jaiswal', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 23.5}, {'Player': 'Rajat Patidar', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 18.0}, {'Player': 'Krunal Pandya', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 3.75}, {'Player': 'Jitesh Sharma', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 1.5}, {'Player': 'Moeen Ali', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 0.2}, {'Player': 'Mohsin Khan', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 0.5}], 'Mahesh': [{'Player': 'Khaleel Ahmed', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Deepak Hooda', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 1.5}, {'Player': 'Karun Nair', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Suryansh Shedge', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 0.1}, {'Player': 'Kagiso Rabada', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 7.0}, {'Player': 'Sanju Samson', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 18.0}, {'Player': 'Akash Madhwal', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 0.1}, {'Player': 'Riyan Parag', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 14.5}, {'Player': 'Vaibhav Suryavanshi', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 0.2}, {'Player': 'Harshal Patel', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 10.0}, {'Player': 'Liam Livingstone', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 15.5}, {'Player': 'Phil Salt', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 15.0}, {'Player': 'Suyash Sharma', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 1.0}, {'Player': 'Akash Deep', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 1.1}, {'Player': 'Nicholas Pooran', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 14.5}, {'Player': 'Abdul Samad', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 0.5}], 'Sandeep': [{'Player': 'Sam Curran', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 2.0}, {'Player': 'Shivam Dube', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 11.0}, {'Player': 'Hardik Pandya', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 20.0}, {'Player': 'Will Jacks', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 3.25}, {'Player': 'Axar Patel', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 19.0}, {'Player': 'T Natarajan', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 5.0}, {'Player': 'Harpreet Brar', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Marco Jansen', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 1.5}, {'Player': 'Azmatullah Omarzai', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 2.0}, {'Player': 'Glenn Phillips', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 5.0}, {'Player': 'Sai Sudharsan', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 14.0}, {'Player': 'Dhruv Jurel', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 1.75}, {'Player': 'Rasikh Salam', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 2.0}, {'Player': 'Angkrish Raghuvanshi', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 1.5}, {'Player': 'Harshit Rana', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 7.0}, {'Player': 'David Miller', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 2.25}], 'Satya': [{'Player': 'Nathan Ellis', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 0.2}, {'Player': 'Matheesha Pathirana', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 14.5}, {'Player': 'Jasprit Bumrah', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 5.25}, {'Player': 'Robin Minz', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 0.0}, {'Player': 'Rohit Sharma', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 13.5}, {'Player': 'Prabhsimran Singh', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 6.25}, {'Player': 'Nehal Wadhera', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Rashid Khan', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 10.5}, {'Player': 'Sai Kishore', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 0.1}, {'Player': 'Heinrich Klaasen', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 17.5}, {'Player': 'Wiaan Mulder', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 0.2}, {'Player': 'Ajinkya Rahane', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 4.0}, {'Player': 'Sunil Narine', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 20.5}, {'Player': 'Manish Pandey', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 0.6}, {'Player': 'Ravi Bishnoi', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 3.0}, {'Player': 'Avesh Khan', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 3.25}], 'Hemanth': [{'Player': 'Ravindra Jadeja', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 10.5}, {'Player': 'KL Rahul', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 27.0}, {'Player': 'Sameer Rizvi', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 1.5}, {'Player': 'Mitchell Starc', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 6.5}, {'Player': 'Tristan Stubbs', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 6.75}, {'Player': 'Arshdeep Singh', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 17.0}, {'Player': 'Yash Thakur', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 0.1}, {'Player': 'Jofra Archer', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 6.75}, {'Player': 'Rahul Chahar', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Swapnil Singh', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 0.1}, {'Player': 'Quinton de Kock', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 7.0}, {'Player': 'Rahmanullah Gurbaz', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 4.25}, {'Player': 'Venkatesh Iyer', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 11.0}, {'Player': 'Ayush Badoni', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 1.0}], 'Yaswanth': [{'Player': 'Rachin Ravindra', 'IPL Team': 'CSK', 'Fantasy Points': 0, 'Price': 15.5}, {'Player': 'Karn Sharma', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 0.2}, {'Player': 'Suryakumar Yadav', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 14.5}, {'Player': 'Shubman Gill', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 20.5}, {'Player': 'Mohammed Siraj', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 4.5}, {'Player': 'Washington Sundar', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 3.0}, {'Player': 'Rahul Tewatia', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Tushar Deshpande', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 0.8}, {'Player': 'Ishan Kishan', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 10.5}, {'Player': 'Abhinav Manohar', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 3.25}, {'Player': 'Kamindu Mendis', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 1.25}, {'Player': 'Mohammed Shami', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 10.0}, {'Player': 'Rinku Singh', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 5.5}, {'Player': 'Shahbaz Ahmed', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Mitchell Marsh', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 2.0}, {'Player': 'Mayank Yadav', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 0.5}], 'Hanish': [{'Player': 'Naman Dhir', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 1.0}, {'Player': 'Mukesh Kumar', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Shreyas Iyer', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 18.5}, {'Player': 'Priyansh Arya', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 3.25}, {'Player': 'Yuzvendra Chahal', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 13.5}, {'Player': 'Jos Buttler', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 17.5}, {'Player': 'Mahipal Lomror', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 0.75}, {'Player': 'Shubham Dubey', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 0.75}, {'Player': 'Wanindu Hasaranga', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 1.5}, {'Player': 'Sandeep Sharma', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 2.25}, {'Player': 'Nitish Kumar Reddy', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 12.0}, {'Player': 'Abhishek Sharma', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 20.5}, {'Player': 'Bhuvneshwar Kumar', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 4.25}, {'Player': 'Vaibhav Arora', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 1.75}, {'Player': 'Ramandeep Singh', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Aiden Markram', 'IPL Team': 'LSG', 'Fantasy Points': 0, 'Price': 1.25}], 'Malli': [{'Player': 'Ryan Rickelton', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 6.75}, {'Player': 'Krishnan Shrijith', 'IPL Team': 'MI', 'Fantasy Points': 0, 'Price': 0.3}, {'Player': 'Jake Fraser-McGurk', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 6.25}, {'Player': 'Abishek Porel', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 6.0}, {'Player': 'Ashutosh Sharma', 'IPL Team': 'DC', 'Fantasy Points': 0, 'Price': 1.0}, {'Player': 'Glenn Maxwell', 'IPL Team': 'PBKS', 'Fantasy Points': 0, 'Price': 8.75}, {'Player': 'Prasidh Krishna', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Ishant Sharma', 'IPL Team': 'GT', 'Fantasy Points': 0, 'Price': 0.5}, {'Player': 'Nitish Rana', 'IPL Team': 'RR', 'Fantasy Points': 0, 'Price': 3.25}, {'Player': 'Pat Cummins', 'IPL Team': 'SRH', 'Fantasy Points': 0, 'Price': 7.25}, {'Player': 'Virat Kohli', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 20.5}, {'Player': 'Tim David', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 2.25}, {'Player': 'Yash Dayal', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 1.25}, {'Player': 'Devdutt Padikkal', 'IPL Team': 'RCB', 'Fantasy Points': 0, 'Price': 1.0}, {'Player': 'Varun Chakravarthy', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 15.0}, {'Player': 'Andre Russell', 'IPL Team': 'KKR', 'Fantasy Points': 0, 'Price': 17.5}]}

for owner, players in team_dict.items():
    for player in players:
        player['Gamedaypoints'] = 0

#print(team_dict)

# Step 1: Fetch Tour Fixtures
fixtures_api_url = "https://fantasy.iplt20.com/classic/api/feed/tour-fixtures?lang=en"
response = requests.get(fixtures_api_url)


fixtures_data = response.json()

# Ensure we are accessing the correct structure
matches = fixtures_data.get("Data", {}).get("Value", [])

# Step 2: Find Today's Match
today_date = datetime.today().strftime("%Y-%m-%d")  # Format: YYYY-MM-DD
#today_date = "2025-03-26"

#print(f"Today's Date: {today_date}")  # Debugging output

fixtures_data = fixtures_data["Data"]["Value"]

today_matches = [match for match in fixtures_data if match["Matchdate"] == today_date]

teams_playing_today = set()

for match in today_matches:
    teams_playing_today.add(match["HomeTeamShortName"])
    teams_playing_today.add(match["AwayTeamShortName"])
    teams_playing_today.add(match["Venue"])

#print("Today's Matches:", today_matches)
#print("Teams Playing Today:", teams_playing_today)

'''for match in matches:
  match_date = match.get("Matchdate")  # Extract Matchdate
            #print(f"Checking Matchdate: {match_date}")  # Debugging step

  if match_date and match_date == today_date:
    todays_match = match
    break'''

tour_gameday_id = max(match["TourGamedayId"] for match in today_matches)#todays_match.get("TourGamedayId")

#print(tour_gameday_id)
#team_gameday_id = todays_match.get("TeamGamedayId")
#home_team = todays_match.get("HomeTeamShortName", "Unknown")
#away_team = todays_match.get("AwayTeamShortName", "Unknown")
#venue = todays_match.get("Venue", "Unknown")

#print(f"Today's Match: {home_team} vs {away_team} | Venue: {venue}")

# Step 3: Fetch Player Data for Today's Match
players_api_url = f"https://fantasy.iplt20.com/classic/api/feed/gamedayplayers?lang=en&tourgamedayId={tour_gameday_id}"
players_response = requests.get(players_api_url)
data = players_response.json()


import json

# Load JSON data from file
#with open(response.json(), "r") as f:
    #data = json.load(f)

# Extract player stats from JSON (Player Name -> Fantasy Points)
players_stats = {
    player["Name"]: player["OverallPoints"]
    for player in data["Data"]["Value"]["Players"]
}

players_stats2 = {
    player["Name"]: player["GamedayPoints"]
    for player in data["Data"]["Value"]["Players"]
}



# Update `Fantasy Points` in `ownerteams`
for owner, players in team_dict.items():
    for player in players:
        player_name = player["Player"]
        if player_name in players_stats:  # If player is in the updated JSON data
            player["Fantasy Points"] = players_stats[player_name]  # Update points
        if player_name in players_stats2:  # If player is in the updated JSON data
            player["Gamedaypoints"] = players_stats2[player_name]  # Update points


# Print updated dictionary
print(team_dict)


RESET = "\033[0m"
BOLD = "\033[1m"
CYAN = "\033[96m"
YELLOW = "\033[93m"
GREEN = "\033[92m"
RED = "\033[91m"
BLUE = "\033[94m"
MAGENTA = "\033[95m"



from datetime import datetime
import pytz

# Set timezone to IST
ist = pytz.timezone('Asia/Kolkata')
now = datetime.now(ist).strftime("%Y-%m-%d %H:%M:%S")

# Print update message
print(f"🟢 Updated on: {now} IST\n")

print(f"{BLUE}{BOLD}LEADERBOARD (by top 11 players){RESET}")

# Calculate total and top 11 points for ranking
owner_scores = []

for owner, players in team_dict.items():
    sorted_players = sorted(players, key=lambda x: int(x['Fantasy Points']), reverse=True)
    total_points = sum(int(player['Fantasy Points']) for player in sorted_players)
    top_11_points = sum(int(player['Fantasy Points']) for player in sorted_players[:11])

    owner_scores.append({
        'Owner': owner,
        'Total Points': total_points,
        'Top 11 Points': top_11_points
    })

# Rank owners by top 11 points (Descending Order)
owner_scores = sorted(owner_scores, key=lambda x: x['Top 11 Points'], reverse=True)

# Assign rank
for rank, owner in enumerate(owner_scores, start=1):
    owner['Rank'] = rank

print(f"\n\033[1mRank     Owner      Total Points    Top 11 Points\033[0m{RESET}")
print("-" * 50)

for owner in owner_scores:
    print(f" {owner['Rank']:<4}   {owner['Owner']:<15} {owner['Total Points']:<12}  {owner['Top 11 Points']:<12}")
    #print("-" * 50)

print("*" * 50)

for match in today_matches:
    home_team = match.get("HomeTeamShortName", "Unknown")
    away_team = match.get("AwayTeamShortName", "Unknown")
    venue = match.get("Venue","Unknown")
    print(f"\n{BOLD}Today's Match(s):{RESET} {home_team} vs {away_team} | Venue: {venue}")

#print(f"\n{BOLD}Today's Match:{RESET} {home_team} vs {away_team} | Venue: {venue}")

print(f"\n{BOLD}{GREEN}Players in Action (for Today){RESET}{RESET}")


owner_wise_data = {}  # Dictionary to store owner-wise players and scores

# Check if any player's IPL team matches the home/away team
for owner, players in team_dict.items():
  owner_players = [ player for player in players if player["IPL Team"] in teams_playing_today ]

  if owner_players:
    total_points = sum(player["Fantasy Points"] for player in owner_players)
    total_gameday_points = sum(player.get("Gamedaypoints", 0) for player in owner_players)

    owner_scores_dict = {entry['Owner']: entry for entry in owner_scores}
    top_11_points = owner_scores_dict.get(owner, {}).get("Top 11 Points", 0)  # Get Top 11 Points from owner_score

    owner_wise_data[owner] = {
                        "players": sorted(owner_players, key=lambda x: x["Fantasy Points"], reverse=True),
                        "total_points": total_points,
                        "top_11_points": top_11_points,
                        "total_gameday_points": total_gameday_points,

                    }

# Print owner-wise data sorted by Top 11 Points
#if owner_wise_data:
sorted_owners = sorted(owner_wise_data.items(), key=lambda x: x[1]["top_11_points"], reverse=True)

for owner, details in sorted_owners:
  print(f"\n{BOLD}Owner: {BLUE}{owner}{RESET}{RESET}")
  for player in details['players']:
    gameday_points = player.get("Gamedaypoints", 0)
    print(f"  - {player['Player']} ({player['IPL Team']})  (Gameday Points: {gameday_points})")
'''else:
  print("\nNo players from your list are playing today.")'''


'''owner_wise_data = {}  # Dictionary to group players by owner

# Check if any player's IPL team matches the home/away team
for owner, players in team_dict.items():
    owner_players = [player["Player"] for player in players if player["IPL Team"] in [home_team, away_team]]

    if owner_players:
      owner_wise_data[owner] = owner_players

# Print owner-wise data
if owner_wise_data:
  for owner, player_list in owner_wise_data.items():
    print(f"\nOwner: {owner}")
    for player in player_list:
      print(f"  - {player}")
else:
  print("\nNo players from your list are playing today.")'''


print("\n")
print("-" * 50)

print("\n\n\033[1mDetailed Player Breakdown:\033[0m")
print("*" * 70)

# Print owner details in order of rank
for owner in owner_scores:
    owner_name = owner['Owner']
    players = sorted(team_dict[owner_name], key=lambda x: int(x['Fantasy Points']), reverse=True)

    print(f"\033[1m{BLUE}{owner['Rank']} {owner_name} (Total Points: {owner['Total Points']}, Top 11 Points: {owner['Top 11 Points']})\033[0m{RESET}")
    print(f"{BOLD}{MAGENTA}           {'Player Name':<14}        {'Team':<4}     {'Fantasy Points':<5}   {'Price(in cr)':<10}{RESET}")
    print("    " + "-" * 65)

    for i, player in enumerate(players, start=1):
        print(f"      {i:<3}  {player['Player']:<20}   {player['IPL Team']:<4}         {player['Fantasy Points']:<10}    {player['Price']:<10}")

    print("*" * 70)
    print("\n")





now = datetime.now(ist).strftime("%Y-%m-%d %H:%M:%S")

# Print update message
print(f"🟢 Updated on: {now} IST\n")

# Flatten the dictionary to get a list of all players with their points
all_players = []
for owner, players in team_dict.items():
    for player in players:
        all_players.append({
            "Owner": owner,
            "Player": player["Player"],
            "Team": player["IPL Team"],
            "Fantasy Points": int(player["Fantasy Points"])  # Convert points to integer for sorting
        })

# Sort the players by fantasy points in descending order
sorted_players = sorted(all_players, key=lambda x: x["Fantasy Points"], reverse=True)

# Print the sorted list
print(f"\n{BOLD}Top Players by Fantasy Points:{RESET}")
print("-" * 40)
for p in sorted_players:
    #print(f"{p['Player']:<10} ({p['Team']}) - {p['Fantasy Points']} points (Owner: {p['Owner']})")
    print(f"   {BLUE} {p['Player']:<20} [{p['Team']:<4}]   {p['Fantasy Points']:<5} (Owner: {p['Owner']})")




import sys


RESET = "\033[0m"
BOLD = "\033[1m"
CYAN = "\033[96m"
YELLOW = "\033[93m"
GREEN = "\033[92m"
RED = "\033[91m"
BLUE = "\033[94m"
MAGENTA = "\033[95m"

import sys
import io
import re

# Mapping ANSI escape codes to CSS styles
ANSI_TO_CSS = {
    "1": "color: black;font-weight: 700;text-align: center;",
    "31": "color: red;text-align: center;",
    "32": "color: green;text-align: center;",
    "33": "color: yellow;text-align: center;",
    "34": "color: blue;text-align: center;",
    "35": "color: magenta;text-align: center;",
    "36": "color: cyan;text-align: center;",
    "37": "color: white;text-align: center;",
    "90": "color: gray;text-align: center;",
    "91": "color: coral;text-align: center;",
    "92": "color: green;text-align: center;",
    "93": "color: yellow;text-align: center;",
    "94": "color: blue;text-align: center;",
    "95": "color: #FF00FF;text-align: center;",
    "96": "color: cyan;text-align: center;"
}

def ansi_to_html(text):
    """Convert ANSI color codes to HTML <span> elements."""
    ansi_pattern = re.compile(r"\033\[(\d+)m")  # Match ANSI escape codes

    def replace(match):
        code = match.group(1)
        if code in ANSI_TO_CSS:
            return f'<span style="{ANSI_TO_CSS[code]}">'
        elif code == "0":  # Reset
            return "</span>"
        return ""

    return ansi_pattern.sub(replace, text) + "</span>"  # Ensure closing tag

class CapturePrintOutput:
    def __enter__(self):
        self.old_stdout = sys.stdout
        sys.stdout = self.buffer = io.StringIO()
        return self

    def __exit__(self, exc_type, exc_value, traceback):
        sys.stdout = self.old_stdout
        output_text = self.buffer.getvalue()
        output_text = ansi_to_html(output_text)  # Convert colors

        html_content = f"""
        <html>
        <body style = "text-align: center;align-items: center;">
            <pre>{output_text}</pre>
        </body>
        </html>
        """

        with open("index.html", "w", encoding="utf-8") as file:
            file.write(html_content)

# Capture print output and save as HTML
with CapturePrintOutput():

    #now = datetime.now(ist).strftime("%Y-%m-%d %H:%M:%S")
    now = datetime.now(ist).strftime("%d-%b-%Y %H:%M:%S")

    # Print update message
    print(f"🟢 Updated on: {now} IST\n")

    print(f"{BLUE}{BOLD}LEADERBOARD (by top 11 points){RESET}{RESET}")

    print(f"\n{BOLD}Rank     Owner      Total Points    Top 11 Points{RESET}")
    print("-" * 50)


    for owner in owner_scores:
      print(f" {owner['Rank']:<4}   {owner['Owner']:<15} {owner['Total Points']:<12}  {owner['Top 11 Points']:<12}")
    print("*" * 50)


    '''for match in today_matches:
      home_team = match.get("HomeTeamShortName", "Unknown")
      away_team = match.get("AwayTeamShortName", "Unknown")
      venue = match.get("Venue","Unknown")
      print(f"\n{BOLD}Today's Match(s):{RESET} {home_team} vs {away_team} | Venue: {venue}")'''



    print(f"\n{BOLD}{MAGENTA}Today's Match(s){RESET}{RESET}")
    print(f"\n{BOLD}{'':<25}{'Teams':<20} {'Venue':<50}{RESET}")
    print("-" * 70)

    for match in today_matches:
      home_team = match.get("HomeTeamShortName", "Unknown")
      away_team = match.get("AwayTeamShortName", "Unknown")
      venue = match.get("Venue", "Unknown")
      print(f"{'':<25}{home_team} vs {away_team:<15} {venue:<50}")

    '''print(f"\n{BOLD}Today's Match:{RESET} {home_team} vs {away_team} | Venue: {venue}")'''

    print("-" * 70)
    print(f"\n{BOLD}{GREEN}Players in Action (for Today){RESET}{RESET}")


    '''for owner, details in sorted_owners:
      print(f"\n{BOLD}Owner: {BLUE}{owner}{RESET}{RESET}")
      for player in details['players']:
        print(f"  - {player['Player']:<20} ({player['IPL Team']:<4}) ({player['gameday_points']})")'''

    for owner, details in sorted_owners:
      print(f"\n{BOLD}Owner: {BLUE}{owner}{RESET}{RESET}")
      for player in details['players']:
        gameday_points = player.get("Gamedaypoints", 0)
        print(f"  - {player['Player']:<25}({player['IPL Team']:<4})(Gameday Points: {gameday_points:<5})")



    print("\n")
    print("-" * 70)

    print("\n\n\033[1mDetailed Player Breakdown:\033[0m")
    print("*" * 70)

    # Print owner details in order of rank
    for owner in owner_scores:
        owner_name = owner['Owner']
        players = sorted(team_dict[owner_name], key=lambda x: int(x['Fantasy Points']), reverse=True)

        print(f"\033[1m{BLUE}{owner['Rank']}. {owner_name} (Total Points: {owner['Total Points']}, Top 11 Points: {owner['Top 11 Points']})\033[0m{RESET}{RESET}")
        print(f"\n{BOLD}{MAGENTA}           {'Player Name':<14}        {'Team':<4}     {'Fantasy Points':<5}   {'Price(in cr)':<10}{RESET}{RESET}")
        print("    " + "-" * 65)

        for i, player in enumerate(players, start=1):
            print(f"      {i:<3}  {player['Player']:<20}   {player['IPL Team']:<4}         {player['Fantasy Points']:<10}    {player['Price']:<10}")

        print("*" * 70)
        print("\n")


    print(f"\n{BOLD}Top Players by Fantasy Points:{RESET}")
    print("-" * 70)
    for p in sorted_players:
      #print(f"{p['Player']:<10} ({p['Team']}) - {p['Fantasy Points']} points (Owner: {p['Owner']})")
      print(f"{BLUE}{p['Player']:<20} [{p['Team']:<4}]   {p['Fantasy Points']:<5} (Owner: {p['Owner']:<10}){RESET}")


print("Output saved to index.html")
```
